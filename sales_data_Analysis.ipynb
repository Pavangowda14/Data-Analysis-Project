```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Diwali_Sales_Data.csv",encoding="unicode_escape")
```

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

Most of the buyers are female

```python
sns.countplot(x="Age Group",data=df,hue="Gender")
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum()
sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

The most of the buyers are of age group betwwen 26-35 years 

```python
state_sales=df.groupby(["State"],as_index=False)["Orders"].sum()
plt.figure(figsize=(15,8))
state_labels=[state[:3] for state in state_sales["State"]]
ax=sns.barplot(x="State",y="Orders",data=state_sales)
ax.set_xticks(range(len(state_labels)))
ax.set_xticklabels(state_labels)
plt.show()
```

```python
df.columns
```

```python
top10_states_sales=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
plt.figure(figsize=(16,8))
ax=sns.barplot(x="State",y="Orders",data=top10_states_sales)
ax.set_title("top 10 states with most number of orders",fontsize=18)
ax.set_xlabel("Top 10 states",fontsize=12)
ax.set_ylabel("Total Orders",fontsize=12)
plt.show()
```

Uttar Pradesh, Maharashtra, Karnataka, Delhi, Madhya Pradesh are the top 5 states with most number of orders	

```python
occupation_sales=df.groupby(["Occupation","Marital_Status"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
plt.figure(figsize=(15,7))
ax=sns.barplot(data=occupation_sales,x="Occupation",y="Orders",hue="Marital_Status")
```

Healthcare, IT sector, Aviation, Banking, and Government are the five occupational sectors that have the highest number of orders and 
People who are not married have made more purchases than married individuals.

```python
product_cat_sales_2=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
plt.figure(figsize=(33,15))
sns.barplot(data=product_cat_sales_2,x="Product_Category",y="Amount")
```

Food, Clothing, Electronic gadgets, Footwears, Furniture are the top 5 product Categories with highest number of purchases
